```python
it_keywords = [
        'developer', 'разработчик', 'программист', 'engineer', 'инженер',
        'backend', 'frontend', 'fullstack', 'full-stack', 'full stack',
        'python', 'java', 'javascript', 'js', 'c++', 'c#', 'php', 'ruby',
        'go', 'rust', 'scala', 'kotlin', 'swift', 'typescript', 'react',
        'angular', 'vue', 'node', 'django', 'flask', 'spring',
        'data scientist', 'data science', 'дата сайентист',
        'data analyst', 'дата аналитик', 'аналитик данных',
        'data engineer', 'дата инженер',
        'ml engineer', 'machine learning', 'машинное обучение',
        'ai engineer', 'artificial intelligence', 'искусственный интеллект',
        'nlp engineer', 'computer vision', 'cv engineer',
        'devops', 'site reliability', 'sre',
        'system administrator', 'системный администратор', 'sysadmin',
        'network engineer', 'сетевой инженер',
        'database administrator', 'dba', 'администратор баз',
        'cloud engineer', 'aws', 'azure', 'gcp',
        'kubernetes', 'docker', 'ci/cd',
        'qa engineer', 'qa инженер', 'тестировщик',
        'test engineer', 'automation engineer', 'автотестировщик',
        'manual qa', 'manual tester',
        'security engineer', 'информационная безопасность',
        'cybersecurity', 'soc analyst', 'pentester',
        'product manager', 'product owner',
        'project manager', 'it project',
        'technical writer', 'технический писатель',
        'business analyst', 'бизнес-аналитик', 'системный аналитик',
        'system analyst',
        'embedded', 'firmware', 'microcontroller',
        'game developer', 'gamedev', 'unity', 'unreal',
        'blockchain', 'smart contract',
        'mobile developer', 'ios developer', 'android developer',
        'tech lead', 'team lead', 'technical lead',
        'architect', 'архитектор', 'solution architect',
        'scrum master', 'agile coach'
    ]

non_it_keywords = [
    'инженер',
    'продаж', 'менеджер по продажам',
    'торговый представитель',
    'маркетолог', 'smm', 'seo',
    'media buyer', 'реклама',
    'бухгалтер', 'экономист', 'финансовый',
    'юрист', 'адвокат',
    'кадров', 'hr', 'рекрутер',
    'водитель', 'курьер',
    'уборщик', 'клининг',
    'продавец', 'кассир',
    'официант', 'повар',
    'врач', 'медсестра',
    'учитель', 'преподаватель'
]
```

```python
def classify_vacancy(title):
    title_lower = title.lower().strip()

    for kw in non_it_keywords:
        if kw in title_lower:
            return False

    for kw in it_keywords:
        if kw in title_lower:
            return True

    return False
```

```python
df['Is_It'] = df['title'].apply(classify_vacancy)
```

```python
df_it = df[df['Is_It']].copy()
```

```python
df_it.drop("Is_It", axis=1, inplace=True)
```

```python
df_it['detailed_information'] = df_it['detailed_information'].fillna("Не указано")
```

```python
df_it = df_it[df_it['salary'] != "Не указано"]
```

```python
def extract_salary_info(salary_info):
    words = salary_info.lower().strip().split()

    currency = None
    for word in words:
        if '₽' in word:
            currency = 'RUB'
            break
        elif '$' in word:
            currency = 'USD'
            break
        elif '€' in word:
            currency = 'EUR'
            break
        elif '₸' in word:
            currency = 'KZT'
            break
        elif "so'm" in word:
            currency = 'SOM'
            break
        elif 'br' in word:
            currency = 'BYN'
            break

    numbers = []
    for word in words:
        clean = word.replace('₽', '').replace('$', '').replace('€', '').replace('₸', '').replace("so'm", '').replace("Br", '').replace(',', '')
        if clean.isdigit():
            numbers.append(int(clean))

    if not numbers:
        return {'amount': None, 'currency': currency}

    amount = sum(numbers) / len(numbers)

    return {'amount': amount, 'currency': currency}
```

```python
salary_info = df_it['salary'].apply(extract_salary_info)
```

```python
df_it['Salary_Amount'] = salary_info.apply(lambda x: x['amount'])
df_it['Salary_Currency'] = salary_info.apply(lambda x: x['currency'])
```

```python
def transfer_to_rubles(row):
    if row['Salary_Currency'] == 'USD':
        return row['Salary_Amount'] * 77
    if row['Salary_Currency'] == 'KZT':
        return row['Salary_Amount'] * 0.15
    if row['Salary_Currency'] == 'EUR':
        return row['Salary_Amount'] * 91
    if row['Salary_Currency'] == 'SOM':
        return row['Salary_Amount'] * 0.88
    if row['Salary_Currency'] == 'BYN':
        return row['Salary_Amount'] * 26.7
    return row['Salary_Amount']
```

```python
df_it['Salary_RUB'] = df_it.apply(transfer_to_rubles, axis=1)
```

```python
df_it = df_it[~df_it['salary'].str.contains('зачас')].copy()
```

```python
df_it = df_it[df_it['Salary_RUB'] >= 25000]
```

```python
df_it = df_it[df_it['Salary_RUB'] < 700000]
```

```python
df_it.drop(['Salary_Amount', 'Salary_Currency', 'salary'], axis=1, inplace=True)
```

```python
def process_exp(exp):
    if exp.startswith("Опыт работы:"):
        exp = exp[12:].strip()

    if exp == "не требуется" or exp == "Не указано":
        return "Нет опыта"
    elif exp == "1–3 года":
        return "1-3 года"
    elif exp == "3–6 лет":
        return "3-6 лет"
    elif exp == "более 6 лет":
        return "6+ лет"

    return exp
```

```python
df_it['experience'] = df_it['experience'].apply(process_exp)
```

```python
no_valid_work_format = ['Формат работы: разъездной', 'Формат работы: наместе работодателя, удалённо, гибрид или разъездной', 'Формат работы: гибрид или разъездной']

df_it = df_it[~df_it['work_format'].isin(no_valid_work_format)].copy()
```

```python
def process_work_format(format):
    if 'гибрид' in format:
        return 'Гибрид'
    if 'наместе' in format:
        return 'Офис'
    if 'удалённо' in format:
        return 'Удалённо'
    return "Не указано"
```

```python
df_it['work_format'] = df_it['work_format'].apply(process_work_format)
```

```python
df_it = df_it[~(df_it['common_employment'] == 'Подработка')]
```

```python
def process_employment(emp):
    if 'полная занятость' in emp.lower():
        return 'Полная занятость'
    elif 'частичная занятость' in emp.lower():
        return 'Частичная занятость'
    elif any(x in emp.lower() for x in ['стажировка', 'проект', 'разовое']):
        return 'Стажировка/Проект'
    return 'Не указано'
```

```python
df_it['common_employment'] = df_it['common_employment'].apply(process_employment)
```

```python
def process_schedule(schedule):
    if 'свободный' in schedule:
        return 'Свободный'
    if '5/2' in schedule:
        return 'Классический (5/2)'
    if 'Не указано' in schedule:
        return 'Не указано'
    return "Другие"
```

```python
df_it['work_schedule'] = df_it['work_schedule'].apply(process_schedule)
```

```python
def process_work_hours(hours):
    if hours == "Не указано":
        return "Не указано"

    hours_lower = hours.lower()

    if 'подоговорённости' in hours_lower.replace(' ', ''):
        return 'По договоренности'

    if '8' in hours:
        return '8 часов'
    if '12' in hours:
        return '12 часов'
    if '6' in hours:
        return '6 часов'

    if any(str(x) in hours for x in range(1, 13)):
        return 'Другие'

    return "Не указано"
```

```python
df_it['work_hours'] = df_it['work_hours'].apply(process_work_hours)
```

```python
def process_payments(payment):
    if payment == "Не указано":
        return "Не указано"

    payment_lower = payment.lower()

    if 'два раза' in payment_lower:
        return 'Два раза в месяц'
    if 'раз вмесяц' in payment_lower:
        return 'Раз в месяц'
    if 'раз внеделю' in payment_lower:
        return 'Раз в неделю'
    if 'проект' in payment_lower:
        return 'За проект'

    return "Не указано"
```

```python
df_it['count_payments'] = df_it['count_payments'].apply(process_payments)
```

```python
df_it['skills'] = df_it['skills'].str.strip('{}')
```

```python
def clean_skills(skills_str):
    skills_list = skills_str.split(",")
    skills = [skill.strip('"') for skill in skills_list]
    return skills
```

```python
df_it['skills'] = df_it['skills'].apply(clean_skills)
```
